```python
import pandas as pd
import spotipy
import spotipy.oauth2 as oauth2
from spotipy.oauth2 import SpotifyOAuth
from spotipy.oauth2 import SpotifyClientCredentials
import time
```

```python
df=pd.read_csv("results.csv") #gets results from csv file
related_words=["bakery"]+list(df["topic"]) +list(df["related_word"]) #adds all entries to one list
related_words=list(set(related_words)) #makes sure there no overlap
```

```python
cid="45d772a85b0a4c2681a42696ad3b5ef3"
cis="9e9848f1081e459ebc686f6f62b5902d"
auth_manager=SpotifyClientCredentials(client_id=cid,client_secret=cis)
sp=spotipy.Spotify(auth_manager=auth_manager)
```

```python
word_results=sp.search(q="track:cupcake")
```

```python
len(word_results["tracks"]["items"])
```

```python
df=pd.DataFrame(columns=["title","artist","release_date","popularity","uri","id","explicit","song_preview"]) #add rows of the columns associated with the song features
```

```python
for song in word_results["tracks"]["items"]:
  row=[]
  row.append(song["name"])
  row.append(song["artists"][0]["name"])
  row.append(song["album"]["release_date"])
  row.append(song["popularity"])
  row.append(song["uri"])
  row.append(song["id"])
  row.append(song["explicit"])
  row.append(song["preview_url"])

  # for this specific track id call that audio features thing and get those special features

  df.loc[len(df)]=row
```
